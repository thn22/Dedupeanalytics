```python
#Import the necessary modules
import pandas as pd
from sqlalchemy import create_engine
from functools import reduce
```

```python
#load the csv
csv_file = "resources/world_risk_index.csv"
risk_data_df = pd.read_csv(csv_file)
risk_data_df.head()
```

```python
#Load and read Life expectancy (2015-2019) csv
csv_file = "resources/HLE.csv"
HLE_df = pd.read_csv(csv_file)
HLE_df.head()
```

```python
#Drop unnecessary columns using df.drop
HLE_cleaned = HLE_df.drop(columns = ['1990' , '1991' , '1992' , '1993' , '1994' , '1995' , '1996' , '1997', '2012', '2013', '2014'])
HLE_cleaned.head()
```

```python
#Drop unnecessary columns
HLE_cleaned = HLE_df.drop(columns = ['Country_Code', 'Region' , '1990' , '1991' , '1992' , '1993' , '1994' , '1995' , '1996' , '1997' , '1998' , '1999' , '2000' , '2001' ,'2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010',  '2011' , '2012', '2013', '2014'])
HLE_cleaned.head()
```

```python
#Remove the Subnational level because it is irrelavant to our database
HLE_cleanedmore_df = HLE_cleaned[HLE_cleaned.Level != 'subnational']
HLE_cleanedmore_df.head()
```

```python
#Rename the columns using df.rename appropriately
HLE_cleaned2_df = HLE_cleanedmore_df.rename(columns = {'2015': 'LE_2015', '2016':'LE_2016' ,
                                                         '2017': 'LE_2017' , '2018':'LE_2018' ,'2019' :'LE_2019'})
                                                       
HLE3_df = HLE_cleaned2_df.drop(columns = ['Level'])
HLE3_df
```

```python
#Load in World Happiness 2019 csv
csv_file = "resources/2019.csv"
H19_data_df = pd.read_csv(csv_file)
H19_data_df.head()
```

```python
#Clean the dataset by dropping unnecessary columns using df.drop  
H19_data_cleaned = H19_data_df.drop(columns = ['GDP per capita' , 'Social support' , 'Freedom to make life choices' ,
                                               'Generosity' , 'Perceptions of corruption' , 'Overall rank' ,'Healthy life expectancy'])
H19_data_cleaned
```

```python
#Rename the columns appropriately
H19_data_cleaned_df = H19_data_cleaned.rename(columns = {'Score': 'Happiness_2019',
                                                         'Healthy life expectancy': 'LE 2019' ,
                                                        'Country or region' : 'Country'})
                                                       
H19_data_cleaned_df
```

```python
#Load in World Happiness 2018 csv
csv_file = "resources/2018.csv"
H18_data_df = pd.read_csv(csv_file)
H18_data_df.head()
```

```python
#Clean the dataset by dropping unnecessary columns using df.drop
H18_data_cleaned = H18_data_df.drop(columns = ['GDP per capita' , 'Social support' , 'Freedom to make life choices' , 
                                               'Generosity' , 'Perceptions of corruption' , 'Overall rank' , 'Healthy life expectancy'])
H18_data_cleaned
```

```python
#Rename the columns appropriately
H18_data_cleaned_df = H18_data_cleaned.rename(columns = {'Score': 'Happiness_2018',
                                                         'Healthy life expectancy': 'LE 2018' ,
                                                        'Country or region' : 'Country'})
                                                       
H18_data_cleaned_df
```

```python
#Load in World Happiness 2017 csv
csv_file = "resources/2017.csv"
H17_data_df = pd.read_csv(csv_file)
H17_data_df.head()
```

```python
#Clean the dataset by dropping unnecessary columns using df.drop
H17_data_cleaned = H17_data_df.drop(columns = ['Whisker.high' , 'Whisker.low' , 'Economy..GDP.per.Capita.' , 
                                               'Family' , 'Freedom' , 'Generosity' , 'Trust..Government.Corruption.' , 
                                               'Dystopia.Residual' , 'Happiness.Rank' , 'Health..Life.Expectancy.' ])
H17_data_cleaned
```

```python
#Rename the columns appropriately
H17_data_cleaned_df = H17_data_cleaned.rename(columns = {'Happiness.Score': 'Happiness_2017',
                                                         'Health..Life.Expectancy.': 'LE 2017'})
                                                       
H17_data_cleaned_df
```

```python
#Load in World Happiness 2016 csv
csv_file = "resources/2016.csv"
H16_data_df = pd.read_csv(csv_file)
H16_data_df.head()
```

```python
#Clean the dataset by dropping unnecessary columns using df.drop
H16_data_cleaned = H16_data_df.drop(columns = ['Region' , 'Happiness Rank' , 'Lower Confidence Interval' ,
                                               'Upper Confidence Interval' , 'Economy (GDP per Capita)' , 'Family' ,
                                               'Generosity' , 'Dystopia Residual' , 'Freedom' , 
                                               'Trust (Government Corruption)' , 'Health (Life Expectancy)' ])
H16_data_cleaned
```

```python
#Rename the columns appropriately
H16_data_cleaned_df = H16_data_cleaned.rename(columns = {'Happiness Score': 'Happiness 2016',
                                                         'Health (Life Expectancy)': 'LE_2016'})
                                                       
H16_data_cleaned_df
```

```python
#Load in World Happiness 2015 csv
csv_file = "resources/2015.csv"
H15_data_df = pd.read_csv(csv_file)
H15_data_df.head()
```

```python
#Clean the dataset by dropping unnecessary columns using df.drop
H15_data_cleaned = H15_data_df.drop(columns = ['Region' , 'Happiness Rank' , 'Standard Error' ,
                                               'Economy (GDP per Capita)' , 'Family' , 'Generosity' , 
                                               'Dystopia Residual' , 'Freedom' , 'Trust (Government Corruption)' , 'Health (Life Expectancy)' ])
H15_data_cleaned
```

```python
#Rename the columns appropriately
H15_data_cleaned_df = H15_data_cleaned.rename(columns = {'Happiness Score': 'Happiness_2015',
                                                         'Health (Life Expectancy)': 'LE 2015'})
                                                       
H15_data_cleaned_df
```

```python
#Final merge of all Individual Happiness csv's using the reduce function
dfs = [H15_data_cleaned_df, H16_data_cleaned_df , H17_data_cleaned_df, H18_data_cleaned_df , H19_data_cleaned_df]

final_df = reduce(lambda  left,right: pd.merge(left,right,on=['Country'],
                                            how='outer'), dfs)

final_df
```

```python
#Establish connection to Postgresql Engine
protocol = 'postgresql'
username = 'postgres'
password = 'postgres1'
host = 'localhost'
port = 5432
database_name = 'Dedupeanalytics'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
#Print table names
engine.table_names()
```

```python
#Load happiness data
final_df.to_sql(name='happiness', con=engine, if_exists='append', index=False)
```
