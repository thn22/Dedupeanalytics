```python
import pandas as pd
from sqlalchemy import create_engine
from functools import reduce
```

```python
le_file = "./Resources/Human_life_Expectancy.csv"
le_df = pd.read_csv(le_file)
le_df.head()
```

```python
le_df = le_df[le_df.Level != 'subnational']
le_df
```

```python
le_df.drop(columns=['1990', '1991','1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
            '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', 'Country_Code', 'Region', 'Level'], inplace=True)
le_df.head()
```

```python
cleaned_le_df = le_df.rename(columns={'2015': 'le_2015', '2016':'le_2016' ,
                                        '2017': 'le_2017' , '2018':'le_2018' ,
                                      '2019' :'le_2019', 'Country': 'country'})
cleaned_le_df
```

```python
year1_file = "./Resources/2015.csv"
year1_df = pd.read_csv(year1_file)
year1_df.head()
```

```python
year1_clean_df = year1_df.drop(columns=['Region' , 'Happiness Rank' , 'Standard Error' ,
                                               'Economy (GDP per Capita)' , 'Family' , 'Generosity' , 
                                               'Dystopia Residual' , 'Freedom' , 'Trust (Government Corruption)', 'Health (Life Expectancy)'])
year1_clean_df.head()
```
